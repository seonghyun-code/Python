**1. Hypothetical data**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
mu1, cov1 = np.array([1.5, 1.5]), np.array([[.30, .05],
                                              [.05, .30]])
rv1 = stats.multivariate_normal(mu1, cov1)
x1 = rv1.rvs(60)

mu2, cov2 = np.array([1.5, -1.5]), np.array([[.50, -.08],
                                              [-.08, .50]])
rv2 = stats.multivariate_normal(mu2, cov2)
x2 = rv2.rvs(60)

mu3, cov3 = np.array([-1.5, 1.5]), np.array([[.10, .03],
                                              [.03, .10]])
rv3 = stats.multivariate_normal(mu3, cov3)
x3 = rv3.rvs(60)

mu4, cov4 = np.array([-1.5, -1.5]), np.array([[.80, .50],
                                              [.50, .80]])
rv4 = stats.multivariate_normal(mu4, cov4)
x4 = rv4.rvs(60)
```

```python
xx = np.linspace(-3, 3, 1)
yy = np.linspace(-3, 3, 1)

plt.figure(figsize=(18, 9))
plt.scatter(x1[:, 0], x1[:, 1], s=15, color="black", label="N(mu1, cov1)")
plt.scatter(x2[:, 0], x2[:, 1], s=15, color="red", label="N(mu2, cov2)")
plt.scatter(x3[:, 0], x3[:, 1], s=15, color="blue", label="N(mu3, cov3)")
plt.scatter(x4[:, 0], x4[:, 1], s=15, color="green", label="N(mu4, cov4)")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel("Baseline Anxiety")
plt.ylabel("Baseline Depression")
plt.grid(True, color="gray", alpha=0.5, linestyle="--")
plt.title("Hypothetical EX")
plt.legend()
plt.show()
```

**2. Dirichlet distribution prior**

```python
# alpha1 = np.array([1, 1, 1])
# d1 = stats.dirichlet.rvs(alpha1, 60)

# alpha2 = np.array([.2, .2, .2])
# d2 = stats.dirichlet.rvs(alpha2, 60)

# alpha3 = np.array([2, 7, 17])
# d3 = stats.dirichlet.rvs(alpha3, 60)


alpha1 = [1, 1, 1]
d1 = np.random.dirichlet(alpha1, 200)

alpha2 = [.2, .2, .2]
d2 = np.random.dirichlet(alpha2, 200)

alpha3 = [2, 7, 17]
d3 = np.random.dirichlet(alpha3, 200)
```

```python
fig = plt.figure(figsize=(24, 6))
ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')
# plt.title("Dirichlet(alpha = [1, 1, 1])")

ax1.set_xlim(0, 1)
ax1.set_xlabel('$\pi_1$')
ax1.set_ylim(0, 1)
ax1.set_ylabel('$\pi_2$')
ax1.set_zlim(0, 1)
ax1.set_zlabel('$\pi_3$')
ax1.scatter(d1[:, 0], d1[:, 1], d1[:, 2], s=15, color="blue")
ax1.view_init(30, 30)
ax1.set_title("Dirichlet($\\alpha$ = [1, 1, 1])")

ax2.set_xlim(0, 1)
ax2.set_xlabel('$\pi_1$')
ax2.set_ylim(0, 1)
ax2.set_ylabel('$\pi_2$')
ax2.set_zlim(0, 1)
ax2.set_zlabel('$\pi_3$')
ax2.scatter(d2[:, 0], d2[:, 1], d2[:, 2], s=15, color="blue")
ax2.view_init(30, 30)
# ax2.set_title("Dirichlet(\\alpha= [0.2, 0.2, 0.2])")
ax2.set_title("Dirichlet($\\alpha$ = [0.2, 0.2, 0.2])")

ax3.set_xlim(0, 1)
ax3.set_xlabel('$\pi_1$')
ax3.set_ylim(0, 1)
ax3.set_ylabel('$\pi_2$')
ax3.set_zlim(0, 1)
ax3.set_zlabel('$\pi_3$')
ax3.scatter(d3[:, 0], d3[:, 1], d3[:, 2], s=15, color="blue")
ax3.view_init(30, 30)
ax3.set_title("Dirichlet($\\alpha$ = [2, 7, 17])")
```
